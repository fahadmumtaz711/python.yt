```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Load the dataset
df_stock = pd.read_csv('D:datasets\Starbucks_dataset.csv')
df_stock.head()
```

```python
# Describe data
df_stock.describe()

# Plot closing price distribution
plt.title('Closing Price Distribution Plot')
sns.histplot(df_stock['Close'], kde=True)
plt.show()

# Relationship between Date and Closing Price
plt.figure(figsize=(12,6))
plt.plot(pd.to_datetime(df_stock['Date']), df_stock['Close'], color='lightcoral')
plt.title('Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.show()
```

```python
# Convert Date to datetime format and set as index
df_stock['Date'] = pd.to_datetime(df_stock['Date'])
df_stock.set_index('Date', inplace=True)

# Feature engineering: Create lag features
df_stock['Prev_Close'] = df_stock['Close'].shift(1)
df_stock = df_stock.dropna()

# Splitting variables
X = df_stock[['Prev_Close']]  # independent variable
y = df_stock[['Close']]       # dependent variable
```

```python
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Regressor model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Prediction result
y_pred_test = regressor.predict(X_test)     # predicted value of y_test
y_pred_train = regressor.predict(X_train)   # predicted value of y_train

# Plot predictions on test set
plt.figure(figsize=(12,6))
plt.scatter(X_test, y_test, color='lightcoral', label='Actual Price')
plt.plot(X_test, y_pred_test, color='firebrick', linewidth=2, label='Predicted Price')
plt.title('Closing Price Prediction (Test Set)')
plt.xlabel('Previous Day Closing Price')
plt.ylabel('Closing Price')
plt.legend()
plt.grid(True)
plt.show()

# Regressor coefficients and intercept
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Evaluate the model
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```
